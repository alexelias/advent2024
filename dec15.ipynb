```python
# part 1

from typing import NamedTuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [up, right, down, left]

def charToDir(char):
    if (char == "<"): return left
    if (char == ">"): return right
    if (char == "^"): return up
    if (char == "v"): return down
    raise NotImplementedError

mapMode = True
grid = []
moves = []
for line in input.splitlines():
    if (line == ""):
        mapMode = False
        continue
    if (mapMode):
        grid.append(list(line))
    else:
        moves += list(line)

width = len(grid[0])
height = len(grid)

#print (grid, moves)

def gridGet(pos):
    return grid[pos.y][pos.x]

def isMovePossible(pos, dir):
    nextTile = gridGet(pos + dir)
    if (nextTile == "#"): return False
    if (nextTile == "."): return True
    if (nextTile == "O"): return isMovePossible(pos + dir, dir)
    raise NotImplementedError

def moveObject(pos, dir):
    if (not isMovePossible(pos, dir)): return False
    nextPos = pos + dir
    if (gridGet(nextPos) == "O"): moveObject(nextPos, dir)
    if (gridGet(nextPos) == "."):
        curTile = gridGet(pos)
        grid[pos.y][pos.x] = "."
        grid[nextPos.y][nextPos.x] = curTile
        return True
    raise NotImplementedError

def findRobot():
    for y in range(height):
        for x in range(width):
            pos = Point(x, y)
            if (gridGet(pos) == "@"): return pos
    raise NotImplementedError

for m in moves:
    dir = charToDir(m)
    moveObject(findRobot(), dir)

boxsum = 0
for y in range(height):
    for x in range(width):
        pos = Point(x, y)
        if (gridGet(pos) == "O"):
            boxsum += (y * 100) + x

print (boxsum)
```

```python
# part 2

from typing import NamedTuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [up, right, down, left]

def isVertical(dir):
    return dir == up or dir == down

def charToDir(char):
    if (char == "<"): return left
    if (char == ">"): return right
    if (char == "^"): return up
    if (char == "v"): return down
    raise NotImplementedError

def doubleTile(tile):
    if (tile == "#"): return "##"
    if (tile == "O"): return "[]"
    if (tile == "."): return ".."
    if (tile == "@"): return "@."
    raise NotImplementedError

mapMode = True
grid = []
moves = []
for line in input.splitlines():
    if (line == ""):
        mapMode = False
        continue
    if (mapMode):
        splitLine = []
        for tile in list(line):
            for nextTile in doubleTile(tile):
                splitLine.append(nextTile)
        grid.append(splitLine)
    else:
        moves += list(line)

width = len(grid[0])
height = len(grid)

print (grid, moves)

def gridGet(pos):
    return grid[pos.y][pos.x]

def oppositeSideDir(tile):
    if (tile == "]"):
        return left
    elif (tile == "["):
        return right
    raise NotImplementedError

def isMovePossible(pos, dir):
#    print (f"isMovePossible {pos}, {dir}")
    nextTile = gridGet(pos + dir)
    if (nextTile == "#"): return False
    if (nextTile == "."): return True
    if (nextTile == "[" or nextTile == "]"):
        if (isVertical(dir)):
           return (isMovePossible(pos + dir, dir) and isMovePossible(pos + dir + oppositeSideDir(nextTile), dir))
        else:
           return isMovePossible(pos + dir, dir)
    print (f"Illegal tile {nextTile} at {pos}, {dir}")
    raise NotImplementedError

def moveObject(pos, dir):
    nextPos = pos + dir
    if (gridGet(nextPos) == "[" or gridGet(nextPos) == "]"):
        if (isVertical(dir)):
            oppositeSide = oppositeSideDir(gridGet(nextPos))
            moveObject(nextPos, dir)
            moveObject(nextPos + oppositeSide, dir)
        else:
            moveObject(nextPos, dir)
    if (gridGet(nextPos) == "."):
        curTile = gridGet(pos)
        grid[pos.y][pos.x] = "."
        grid[nextPos.y][nextPos.x] = curTile
        return True
    raise NotImplementedError

def findRobot():
    for y in range(height):
        for x in range(width):
            pos = Point(x, y)
            if (gridGet(pos) == "@"): return pos
    raise NotImplementedError

for m in moves:
    dir = charToDir(m)
    pos = findRobot()
    if (isMovePossible(pos, dir)):
        moveObject(pos, dir)

boxsum = 0
for y in range(height):
    for x in range(width):
        pos = Point(x, y)
        if (gridGet(pos) == "["):
            boxsum += (y * 100) + x

print (boxsum)
```
