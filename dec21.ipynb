```python
# part 2

from typing import NamedTuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [left, up, down, right]

def dirToChar(dir: Point) -> str:
    if (dir == up): return "^"
    if (dir == down): return "v"
    if (dir == left): return "<"
    if (dir == right): return ">"
    raise NotImplementedError

def manhattanDist(pt1: Point, pt2: Point) -> int:
    return abs(pt2.x - pt1.x) + abs(pt2.y - pt1.y)

grid = dpad.splitlines()
def gridGet(pt: Point) -> str:
    if (oob(pt)): return "#"
    return grid[pt.y][pt.x]

height = len(grid)
width = len(grid[0])

def oob(pt: Point) -> bool:
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)

def findChar(char: str) -> Point:
    if (len(char) > 1): raise NotImplementedError
    for y in range(height):
        for x in range(width):
            pt = Point(x, y)
            if (gridGet(pt) == char):
                return pt
    raise NotImplementedError

from functools import cache 

@cache
def findPathLength(depth: int, targetDepth: int, dstStr: str) -> int:
    if (depth == targetDepth):
        return len(dstStr)    

    ret = 0

    startPt = findChar("A")
    for dstChar in dstStr:
        dstPt = findChar(dstChar)

        path = ""
        for dirindex in range(4):
            firstDistance = 0
            firstDir = directions[dirindex]
            if (firstDir == left and dstPt.x >= startPt.x): continue
            if (firstDir == right and dstPt.x <= startPt.x): continue
            if (firstDir == up and dstPt.y >= startPt.y): continue
            if (firstDir == down and dstPt.y <= startPt.y): continue
            if (firstDir == left or firstDir == right):
                firstDistance = abs(dstPt.x - startPt.x)
            if (firstDir == up or firstDir == down):
                firstDistance = abs(dstPt.y - startPt.y)
            firstWallPt = startPt + firstDir * firstDistance
            if (gridGet(firstWallPt) == "#"): continue
            path = path + dirToChar(firstDir) * firstDistance
            break
        secondDistance = 0
        if (firstDir == left or firstDir == right):
            secondDistance = abs(dstPt.y - startPt.y)
            secondDir = up
            if (dstPt.y > startPt.y): secondDir = down
        else:
            secondDistance = abs(dstPt.x - startPt.x)
            secondDir = left
            if (dstPt.x > startPt.x): secondDir = right

        path = path + dirToChar(secondDir) * secondDistance
        path = path + "A"

        startPt = dstPt

        ret += findPathLength(depth + 1, targetDepth, path)
    
    return ret

print (f"Desired lengths for 029A: numpath: {len("<A^A>^^AvvvA")}, dirpath1: {len("v<<A>>^A<A>AvA<^AA>A<vAAA>^A")}, dirpath2: {len("<vA<AA>>^AvAA<^A>A<v<A>>^AvA^A<vA>^A<v<A>^A>AAvA^A<v<A>A>^AAAvA<^A>A")}")

complexity = 0
targetDepth = 25
for line, dpadinputs in processedpairs:
    print (f"Computing for {line}, {dpadinputs}...")
    length = findPathLength(0, targetDepth, dpadinputs)
    print (f"For {line} depth {targetDepth}: length {length})")
    
    complexity += length * int(line[:-1])

print (f"complexity: {complexity}")
```
