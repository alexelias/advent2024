```python
input = r"""47|53
97|13
97|61
97|47
75|29
61|13
75|53
29|13
97|29
53|29
61|53
97|53
61|29
47|13
75|47
97|75
47|61
75|61
47|29
75|13
53|13

75,47,61,53,29
97,61,53,29,13
75,29,13
75,97,47,61,53
61,13,29
97,13,75,29,47
"""
```

```python
# Part 1
from typing import List, Dict
from collections import namedtuple

Rule = namedtuple('Rule', ['before', 'after'])

mode = False
rules: List[Rule] = []
prints: List[int] = []
for line in input.splitlines():
    if (len(line) == 0): mode = True; continue

    if (not mode):
        before, after = line.split("|")
        rules.append(Rule(int(before), int(after)))
    else:
        prints.append(list(map(int,line.split(","))))

print(rules)
print(prints)

correctPrintJobIds: List[int] = []
for jobId, printJob in enumerate(prints):
    try:
        for pageIndex, page in enumerate(printJob):
            for rule in rules:
                if (rule.before != page): continue
                for otherPage in printJob[:pageIndex]:
                    if (otherPage == rule.after):
                        raise StopIteration
        correctPrintJobIds.append(jobId)
    except StopIteration:
        pass
                    
print(correctPrintJobIds)

checksum = 0
for id in correctPrintJobIds:
    job = prints[id]
    middleValue = job[(len(job) // 2)] 
    checksum += middleValue

print(checksum)
```

```python
# Part 2
from typing import List, Dict
from collections import namedtuple

Rule = namedtuple('Rule', ['before', 'after'])

mode = False
rules: List[Rule] = []
prints: List[int] = []
for line in input.splitlines():
    if (len(line) == 0): mode = True; continue

    if (not mode):
        before, after = line.split("|")
        rules.append(Rule(int(before), int(after)))
    else:
        prints.append(list(map(int,line.split(","))))

print(rules)
print(prints)

incorrectPrintJobIds: List[int] = []
for jobId, printJob in enumerate(prints):
    try:
        for pageIndex, page in enumerate(printJob):
            for rule in rules:
                if (rule.before != page): continue
                for otherPage in printJob[:pageIndex]:
                    if (otherPage == rule.after):
                        incorrectPrintJobIds.append(jobId)
                        raise StopIteration
    except StopIteration:
        pass
                    
print(incorrectPrintJobIds)

def doOneSwap(printJob):
    for pageIndex, page in enumerate(printJob):
        for rule in rules:
            if (rule.before != page): continue
            for otherPageIndex, otherPage in enumerate(printJob[:pageIndex]):
                if (otherPage == rule.after):
                    printJob[pageIndex] = otherPage
                    printJob[otherPageIndex] = page
                    return True
    return False

for jobId in incorrectPrintJobIds:
    printJob = prints[jobId]
    while doOneSwap(printJob):
        pass

checksum = 0
for id in incorrectPrintJobIds:
    job = prints[id]
    middleValue = job[(len(job) // 2)] 
    checksum += middleValue

print(checksum)
```
