```python
input = r"""....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...
"""
```

```python
# Part 1
from typing import List, Dict
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])

grid = input.splitlines()

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
def rotateClockwise(dir: Point):
    if (dir == up): return right
    if (dir == right): return down
    if (dir == down): return left
    if (dir == left): return up
    raise NotImplementedError

obstructions = []
walkDir = up
guardPos = Point(0, 0)
for y, row in enumerate(grid):
    height = y + 1
    for x, column in enumerate(row):
        width = x + 1
        if (column == "#"):
           obstructions.append(Point(x, y))
        elif (column == "^"):
            guardPos = Point(x, y)

print(width, height, guardPos, obstructions)

def oob(pt: Point):
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)

while True:
    currentRow = grid[guardPos.y]
    grid[guardPos.y] = currentRow[:guardPos.x] + "X" + currentRow[guardPos.x + 1:]
    print(guardPos.x, guardPos.y, grid[guardPos.y])
    
    nextPos = Point(guardPos.x + walkDir.x, guardPos.y + walkDir.y)
    if (oob(nextPos)):
        break
    if (grid[nextPos.y][nextPos.x] == "#"):
        walkDir = rotateClockwise(walkDir)
        continue
    guardPos = nextPos

print(grid)

xcount = 0
for y, row in enumerate(grid):
    for x, column in enumerate(row):
        if (column == "X"):
            xcount += 1

print(xcount)
```

```python
# Part 2
from typing import List, Dict
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])

grid = input.splitlines()

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
def rotateClockwise(dir: Point):
    if (dir == up): return right
    if (dir == right): return down
    if (dir == down): return left
    if (dir == left): return up
    raise NotImplementedError

obstructions = []
walkDir = up
guardPos = Point(0, 0)
for y, row in enumerate(grid):
    height = y + 1
    for x, column in enumerate(row):
        width = x + 1
        if (column == "#"):
           obstructions.append(Point(x, y))
        elif (column == "^"):
            guardPos = Point(x, y)

print(width, height, guardPos, obstructions)

def oob(pt: Point):
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)

def guardWalk(guardPos, walkDir):
    numSteps = 0
    while numSteps < 10000:
        nextPos = Point(guardPos.x + walkDir.x, guardPos.y + walkDir.y)
        if (oob(nextPos)):
            return False
        if (grid[nextPos.y][nextPos.x] == "#"):
            walkDir = rotateClockwise(walkDir)
            continue
        guardPos = nextPos
        numSteps += 1
    return True

count = 0
for y in range(height):
    for x in range(width):
        currentRow = grid[y]
        if (currentRow[x] != "."): continue
        grid[y] = currentRow[:x] + "#" + currentRow[x + 1:]
        print(x, y, grid[y])
        if (guardWalk(guardPos, walkDir)): count += 1
        grid[y] = currentRow[:x] + "." + currentRow[x + 1:]

print(count)
```
