```python
width = 7
height = 7
limit = 12
input = r"""5,4
4,2
4,5
3,0
2,1
6,3
2,4
1,5
0,6
3,3
2,6
5,1
1,2
5,5
2,5
6,5
1,4
0,4
6,4
1,1
6,1
1,0
0,5
1,6
2,0
"""
```

```python
# part 1

from typing import NamedTuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [up, right, down, left]

grid = []
for y in range(height):
    line = []
    for x in range(width):
        line.append('.')
    grid.append(line)

count = 0
for str in input.splitlines():
    x, y = str.split(',')
    grid[int(y)][int(x)] = '#'
    count += 1
    if (count == limit):
        break

def gridGet(pt):
    return grid[pt.y][pt.x]

def oob(pt: Point):
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)

import networkx as nx
from networkx.algorithms.shortest_paths.weighted import single_source_dijkstra

Edge = NamedTuple('Edge', [('src', Point), ('dst', Point), ('weight', int)])

graph = nx.DiGraph()
for y in range(height):
    for x in range(width):
        pt = Point(x, y)
        if (gridGet(pt) == "#"): continue
        for dirindex in range(4):
            dir = directions[dirindex]
            nextpt = pt + dir
            if (oob(nextpt)):
                continue
            if (gridGet(nextpt) != "#"):
                graph.add_edge(pt, nextpt, weight=1)

startpt = Point(0, 0)
endpt = Point(width - 1, height - 1)
shortest_path = nx.single_source_dijkstra(graph, source=startpt, target=endpt, weight='weight')

print (shortest_path)
```
